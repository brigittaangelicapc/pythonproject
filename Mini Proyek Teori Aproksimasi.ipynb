```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
x=symbols('x')
a=chebyshevt(2,x)+1.05

print("Fungsi Chebyshev dengan penambahan 1.05: ")
print(a)
print("")
print("Dengan menggunakan nilai mutlak 1.2*x, maka:")
print("|1.2*x|")
print("")
print("Maka, dengan menggunakan grafik akan didapat:")

x = np.linspace(-1,1,100)

#Penulisan fungsi
y = abs(1.2*x)
z = 2*x**2-1+1.05

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('none')

#Keterangan untuk bagian grafik
plt.plot(x,y, 'blue', label='|1.2*x|')
plt.plot(x,z, 'black', label='Chebyshev + 1.05')
plt.legend(loc='upper left')
plt.show()
```
